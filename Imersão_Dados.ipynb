## **Desafios da Aula 2**

### **Desafio 01:** Ordenar o gráfico countplot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order = dados.query('composto in @cod_compostos')['composto'].value_counts().index)
ax.set_title('Top 5 compostos')
plt.show()
```

### **Desafio 02:** Melhorar a visualização alterando tamanho da fonte...

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order = dados.query('composto in @cod_compostos')['composto'].value_counts().index)
ax.set_title('Top 5 compostos', weight='bold', fontdict={'fontsize':15})
ax.set_xlabel('Compostos', weight='bold', fontdict={'fontsize':12})
ax.set_ylabel('Quantidade de compostos', weight='bold', fontdict={'fontsize':12})
plt.show()
```

### **Desafio 03:** Plotar os histogramas com seaborn.

```python
ax = sns.histplot(x='g0', data=dados, bins=100, kde=True)
ax.set_title('Histograma da coluna g0', weight='bold', fontdict={'fontsize': 15})
ax.set_xlabel('Intervalo', weight='bold', fontdict={'fontsize': 12})
ax.set_ylabel('Contagem', weight='bold', fontdict={'fontsize': 12})
plt.show()
```

```python
ax = sns.histplot(x=dados.loc[:, 'g0':'g771'].describe().T['mean'], data=dados, bins=30)
ax.set_title("Histograma dos valores médios das colunas G's", weight='bold', fontdict={'fontsize': 12})
ax.set_xlabel('Intervalo', weight='bold', fontdict={'fontsize': 10})
ax.set_ylabel('Contagem', weight='bold', fontdict={'fontsize': 10})
plt.show()
```

```python
ax = sns.histplot(x=dados.loc[:, 'g0':'g771'].describe().T['min'], data=dados, bins=30)
ax.set_title("Histograma dos valores mínimos das colunas G's", weight='bold', fontdict={'fontsize': 12})
ax.set_xlabel('Intervalo', weight='bold', fontdict={'fontsize': 10})
ax.set_ylabel('Contagem', weight='bold', fontdict={'fontsize': 10})
plt.show()
```

```python
ax = sns.histplot(x=dados.loc[:, 'g0':'g771'].describe().T['max'], data=dados, bins=30)
ax.set_title("Histograma dos valores máximos das colunas G's", weight='bold', fontdict={'fontsize': 12})
ax.set_xlabel('Intervalo', weight='bold', fontdict={'fontsize': 10})
ax.set_ylabel('Contagem', weight='bold', fontdict={'fontsize': 10})
plt.show()
```

```python
ax = sns.histplot(x=dados.loc[:, 'c0':'c99'].describe().T['mean'], data=dados, bins=30)
ax.set_title("Histograma dos valores médios das colunas C's", weight='bold', fontdict={'fontsize': 12})
ax.set_xlabel('Intervalo', weight='bold', fontdict={'fontsize': 10})
ax.set_ylabel('Contagem', weight='bold', fontdict={'fontsize': 10})
plt.show()
```
